# 1. 모듈 로드

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision import transforms, datasets
```

# 2. 장치확인

```python
if torch.cuda.is_available():
    DEVICE = torch.device('cuda')
else:
    DEVICE = torch.device('cpu')
print(f'Using Pytorch version: {torch.__version__}, Device: {DEVICE}')
```

# 3. 데이터 load

```python
BATCH_SIZE = 32
EPOCHS = 10
```

```python
# 데이터 다운로드
train_dataset = datasets.CIFAR100(root = "../data/CIFAR100",
                                  train = True,
                                  download = True,
                                  transform = transforms.ToTensor())
test_dataset = datasets.CIFAR100(root = "../data/CIFAR100",
                                 train = False,
                                 transform = transforms.ToTensor())
```

```python
train_loader = torch.utils.data.DataLoader(dataset = train_dataset,
                                           batch_size = BATCH_SIZE,
                                           shuffle = True)
```

# 4. 데이터 확인하기

```python
for (X_train, y_train) in train_loader:
    print(f'X_train size: {X_train.size()}, type: {X_train.type()}')
    print(f'y_train size: {y_train.size()}, type: {y_train.type()}')
    break
```

```python
pltsize = 1
plt.figure(figsize = (10*pltsize, pltsize))
for i in range(10):
    plt.subplot(1, 10, i+1)
    plt.axis('off')
    plt.imshow(np.transpose(X_train[i, :, :, :].numpy(), (1, 2, 0)))
    plt.title('Class ' + str(y_train[i].item()))
plt.show()
```

# 5. 모델 load

```python
from pl_bolts.models.autoencoders import AE

model = AE(input_height=32, enc_type="resnet18")
```

# 6. 학습

```python
from pytorch_lightning import Trainer

trainer = Trainer(gpus=1, max_epochs=5)
trainer.fit(model, train_loader)
```

# 7. 결과

```python
predicted_images = model(X_train)
```

```python
pltsize = 1
plt.figure(figsize = (10*pltsize, pltsize))
for i in range(10):
    plt.subplot(1, 10, i+1)
    plt.axis('off')
    plt.imshow(np.transpose(predicted_images[i, :, :, :].detach().numpy(), (1, 2, 0)))
    plt.title('Class ' + str(y_train[i].item()))
plt.show()
```
