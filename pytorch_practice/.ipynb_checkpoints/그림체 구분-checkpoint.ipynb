### 데이터 전처리

```python
mean_rgb = (0.485, 0.456, 0.406)
std_rgb = (0.229, 0.224, 0.225)

transformer = transforms.Compose([
                transforms.ToTensor(),
                transforms.Normalize(mean_rgb, std_rgb)
])

# 이미지에 transformation 적용하기
dataset = []
for i in range(len(thumbs_byte)):
    if thumbs_byte[i].mode != "RGB":
        thumbs_byte[i] = thumbs_byte[i].convert("RGB")
    #thumbs_byte[i] = thumbs_byte[i].convert("L")
    tmp = transformer(thumbs_byte[i])
    dataset.append(tmp)
```

# 5. Pre-trained Model 불러오기

- VGG19

```python
model = models.vgg19(pretrained=True).features.to(DEVICE).eval()
print(model)
```

```python
feature_layers = {'0': 'conv1_1',
                  '5': 'conv2_1',
                  '10': 'conv3_1',
                  '19': 'conv4_1',
                  '21': 'conv4_2',
                  '28': 'conv5_1'}
```

## 7. Style loss func

```python
def get_features(x, model, layers):
    features = {}
    for name, layer in enumerate(model.children()):
        x = layer(x)
        if str(name) in layers:
            features[layers[str(name)]] = x
    return features
```

```python
def gram_matrix(x):
    n, c, h, w = x.size()
    x = x.view(n*c, h*w)
    gram = torch.mm(x,x.t())
    return gram
```

```python
def get_style_loss(pred_features, target_features, style_layers_dict):
    loss = 0
    for layer in style_layers_dict:
        pred_fea = pred_features[layer]
        pred_gram = gram_matrix(pred_fea)
        n, c, h, w = pred_fea.shape
        target_gram = gram_matrix(target_features[layer])
        layer_loss = style_layers_dict[layer] * F.mse_loss(pred_gram, target_gram)
        loss += layer_loss / (n*c*h*w)
    return loss
```

# 8. style loss 계산

```python
style_loss = np.zeros((len(dataset),len(dataset)))
style_loss.shape
```

```python
style_layers_dict = {'conv1_1':0.75,
                     'conv2_1':0.5,
                     'conv3_1':0.25,
                     'conv4_1':0.25,
                     'conv5_1':0.25}
```

```python
def get_ith_loss(row, style_loss):
    
    sty_tensor = dataset[row].unsqueeze(0).to(DEVICE)
    row_image = get_features(sty_tensor, model, feature_layers)
    for col in range(len(dataset)):
        sty_tensor = dataset[col].unsqueeze(0).to(DEVICE)
        col_image = get_features(sty_tensor, model, feature_layers)
        style_loss[row, col] = get_style_loss(row_image, col_image, style_layers_dict)
    
    torch.cuda.empty_cache()
    
    return style_loss 
```

# 9. 결과

```python
def show_result(row, style_loss):
    style_loss = get_ith_loss(row, style_loss)
    idx = np.argsort(style_loss[row])

    # Best 5
    best5 = idx[1:6]
    # Worst 5
    worst5 = idx[-5:][::-1]
    
    
    plt.axis('off')
    plt.imshow(thumbs_byte[row])
    
    pltsize = 2
    plt.figure(figsize = (5 * pltsize, pltsize))

    # best 5 결과
    for i in range(5):
        plt.subplot(1, 5, i+1)
        plt.axis('off')
        plt.imshow(thumbs_byte[best5[i]])
        plt.title(f'loss:{style_loss[row,best5[i]]:.2f}')
        #plt.title(f'title: {thumbs_path[best5[i]]}')
    
    pltsize = 2
    plt.figure(figsize = (5 * pltsize, pltsize))

    # worst 5 결과
    for i in range(5):
        plt.subplot(1, 5, i+1)
        plt.axis('off')
        plt.imshow(thumbs_byte[worst5[i]])
        plt.title(f'loss:{style_loss[row,worst5[i]]:.2f}')
        #plt.title(f'title: {thumbs_path[worst5[i]]}')
        
    return style_loss
```
