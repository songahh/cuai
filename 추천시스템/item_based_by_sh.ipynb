```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
item_sim = cosine_similarity(ratings_T, ratings_T)
```

```python
item_sim_df = pd.DataFrame(data=item_sim, index=ratings.columns, columns=ratings.columns)
```

```python
def predict_rating(ratings_arr, item_sim_arr):
    ratings_pred = ratings_arr.dot(item_sim_arr) / np.array([np.abs(item_sim_arr).sum(axis=1)])
    return ratings_pred
```

```python
ratings_pred = predict_rating(ratings.values, item_sim_df.values)
ratings_pred_matrix = pd.DataFrame(data=ratings_pred, index=ratings.index, columns=ratings.columns)
ratings_pred_matrix[:3]
```

```python
from sklearn.metrics import mean_squared_error
```

```python
def get_mse(pred, actual):
    pred = pred[actual.nonzero()].flatten()
    actual = actual[actual.nonzero()].flatten()
    return mean_squared_error(pred, actual)
```

```python
print('아이템 기반 모든 최근접 이웃 MSE: ', get_mse(ratings_pred, ratings.values))
```

```python
def predict_rating_topsim(ratings_arr, item_sim_arr, n=20):
    pred = np.zeros(ratings_arr.shape)

    for col in range(ratings_arr.shape[1]):
        top_n_items = [np.argsort(item_sim_arr[:, col])[:-n-1:-1]]
        for row in range(ratings_arr.shape[0]):
            pred[row, col] = item_sim_arr[col, :][top_n_items].dot(ratings_arr[row, :][top_n_items].T)
            pred[row, col] /= np.sum(np.abs(item_sim_arr[col, :][top_n_items]))

    return pred
```

```python
ratings_pred = predict_rating_topsim(ratings.values, item_sim_df.values, n=20)
print('아이템 기반 최근접 TOP-20 이웃 MSE: ', get_mse(ratings_pred, ratings.values))
```

```python
ratings_pred_matrix = pd.DataFrame(data=ratings_pred, index=ratings.index, columns=ratings.columns)
```

```python
def get_unseen_webtoons(ratings, userId):
    user_rating = ratings.loc[userId, :]
    
    already_seen = user_rating[user_rating > 0].index.tolist()

    webtoons_list = ratings.columns.tolist()

    unseen_list = [webtoon for webtoon in webtoons_list if webtoon not in already_seen]

    return unseen_list
```

```python
def recomm_webtoon_by_userid(pred_df, userId, unseen_list, top_n=10):
    recomm_webtoons = pred_df.loc[userId, unseen_list].sort_values(ascending=False)[:top_n]
    return recomm_webtoons
```

```python
unseen_list = get_unseen_webtoons(ratings, 9)
```

```python
recomm_webtoons = recomm_webtoon_by_userid(ratings_pred_matrix, 9, unseen_list, top_n=10)
```

```python
recomm_webtoons = pd.DataFrame(data=recomm_webtoons.values, index=recomm_webtoons.index, columns=['pred_score'])
```
